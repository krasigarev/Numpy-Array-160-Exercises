```python
import numpy as np
```

```python
a = np.array([[1,2,3], [2,3,6], [6,7,2]])
print('Original: \n', a)
```

```python
print('triu() without any parameter: \n', np.triu(a))
```

```python
print('triu() with any parameter: \n', np.triu(a, 1))
```

```python
print('triu() with any parameter: \n', np.triu(a, -1))
```
