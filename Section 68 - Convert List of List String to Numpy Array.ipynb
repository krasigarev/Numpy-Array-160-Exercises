```python
import numpy as np
import re
```

```python
a_list = '''
[[25 3 2]
 [1 21 0]
 [1 0 0]]
'''
print(a_list)
```

```python
rows = []

for line in filter(len, map(str.strip, a_list.split("\n"))):
    rows.append([int(v) for v in re.findall(r"\b[0-9]+\b", line)])

print(rows)
```

```python
np_rows = np.array(rows)
print(np_rows)
```

```python
np_rows += 2
print(np_rows)
```
