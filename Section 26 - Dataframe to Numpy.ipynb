```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'a1': [1,2,3],
    'a2': [4,5,6],
    'a3': [7,8,9]
})

print(df)

x = df.to_numpy()
print('\n DataFrame to numpy array: ') 
print(x)

y = df.index.to_numpy()
print('\n DataFrame indices to Numpy') 
print(y)

z = df['a1'].to_numpy()
print('\n DataFrame series to Numpy')
print(z)
```
