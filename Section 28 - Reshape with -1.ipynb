```python
import numpy as np
```

```python
a = np.matrix([
    [1,2,3,4],
    [2,3,4,5],
    [3,4,5,6]
])

print('Original Numpy array: \n', a)
print('Original Numpy array shape: \n', a.shape)
```

```python
c = np.reshape(a, -1)
print('new shape: ', c)
```

```python
f = np.reshape(a, (2, -1))
print(f)
```

```python
g = np.reshape(a, (3, -1))
print(g)
```

```python
j = np.reshape(a, (4, -1))
print(j)
```
