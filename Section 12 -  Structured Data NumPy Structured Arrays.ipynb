```python
import numpy as np
```

```python
name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = [25,45,37,19]
weight = [55.0,85.5,68.0,61.5]

data = np.zeros(4, dtype={'names':('name', 'age', 'weight'), 'formats':('U10', 'i4', 'f8')})
print(data.dtype)
```

```python
data['name'] = name
data['age'] = age
data['weight'] = weight
print(data)
```
